When to use assign?

    - Add multiple column in a single line  of code
    - When you need to overwrite the values of an existing columns(best practice)



It returns a new object(a copy) with all the original columns in addition to the new ones

# 1 Creating a conditional column for more than two choices: np.select()

```python
# create an array based on multiple new recovered (+2 choices)

# create "Conditions" and "Values" variables

conditions = [
    df_covid['New recovered'] > 2000,
    (df_covid['New recovered'] > 1000) & (df_covid['New recovered']<=2000),
    (df_covid['New recovered'] > 800) & (df_covid['New recovered']<=1000),
    df_covid['New recovered'] <= 500

]


values = ['Very High', 'High', 'Low', 'Very Low']
```

```python
# set it to a new column
df_covid['Affected'] = np.select(conditions, values)
```

```python
# show dataframe 
df_covid
```

```python
# count values
df_covid['Affected'].value_counts()
```

# 3.1 Single Barplot

```python
# convert 'Death' column to numeric
df_covid["Death"] = pd.to_numeric(df_covid["Death"], errors="coerce")
# Group by state and sum deaths
df_state = df_covid.groupby("Name of State / UT")["Death"].sum()

# Sort by deaths (highest first)
df_state_sorted = df_state.sort_values(ascending=False)

# Take only top 10
df_top10 = df_state_sorted.head(10)

# Plot bar
df_top10.plot(kind="bar", color='pink', figsize=(10,5), title="Top 10 States by COVID Deaths")
```

# 3.2 Barplot grouped by multiple variables 

```python
# group by state and sum for multiple columns
df_state_multi = df_covid.groupby('Name of State / UT') [['Death', 'Cured/Discharged/Migrated', 'Total Confirmed cases']].sum()

# take top 5 states by confirmed cases (to keep the graph readable)
df_top5_multi = df_state_multi.sort_values('Total Confirmed cases', ascending=False).head(5)

# plot grouped barplot
df_top5_multi.plot(kind='bar', figsize=(10, 7), title='COVID Comparison (Top 5 States)')
```

```python
# Get top 5 states by total confirmed cases
df_state_multi = df_covid.groupby('Name of State / UT')['Total Confirmed cases'].sum()

top5_states = df_state_multi.sort_values(ascending=False).head(5).index

# Now filter original DataFrame
df_top5_data = df_covid[df_covid['Name of State / UT'].isin(top5_states)]

# Pivot so each state becomes a separate column
df_pivot = df_top5_data.pivot_table(index='Date',
                                    columns='Name of State / UT',
                                    values='Total Confirmed cases')

# Step 4: Boxplot using cufflinks
df_pivot.iplot(kind='box',
               title='Boxplot of Total Confirmed Cases by Top 5 States',
               colors=['red','blue','green','orange','purple'])
```
